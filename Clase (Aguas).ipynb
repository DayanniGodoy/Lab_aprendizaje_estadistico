```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
```

```python
data = pd.read_excel("Variables_cualitativas.xlsx")
data
```

```python
X = data.drop(columns = ["cali"])
y = data["cali"]
```

```python
n = len(y)
ones = np.ones([n,1])
X = np.hstack([ones,X])
```

```python
ols = sm.OLS(y, X)
results = ols.fit()
results.summary()
```
