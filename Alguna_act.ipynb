```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeRegressor
```

1-  Tomar Advertising.csv y hacer una regresión (Linear Regression) como benchmark (R2)

2- Usar un decision tree y ajustar con dataset (R2)

3- Coomparar R2 Escribe observaciones

4- Repetir con train test split (random sgtate = 42)

5- Repetir (1-3) con PolynomialFeatures(degree=2)

```python
data = pd.read_csv('Advertising.csv')
```

```python
X = data[['TV', 'radio', 'newspaper']]
y = data['sales']
```

```python
lr = LinearRegression()
lr.fit(X,y)
b0 = lr.intercept_
print("Intercepto (b0):", b0)  # constante
b1 = lr.coef_
print("Coeficientes (b1):", b1)       # coeficientes
R2_lr = r2_score(y, lr.predict(X)) 
print("R2:", R2_lr)   
```

```python
#Decision Tree
dt = DecisionTreeRegressor()
dt.fit(X,y)
R2_dt= r2_score(y, dt.predict(X))
print("R2 Decision Tree:",R2_dt)   # error
```

### Observación
Holy shit aumentó un montón con el decision tree

## Train Test split

```python
# 4. Repetir con traintest split
# Dividimos el dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,y, 
    test_size=0.20,
    random_state=42,
)

# StandardScaler()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   
X_test_scaled = scaler.transform(X_test) 

# Regresión Lineal con train test split
lr_tt = LinearRegression()
lr_tt.fit(X_train_scaled, y_train) 
R2_lr_tt = r2_score(y_test, lr_tt.predict(X_test_scaled)) 
print("R2 Train Test Split Linear Regression:", R2_lr_tt) 
```

```python
#Decision Tree
dt_tt = DecisionTreeRegressor()
dt_tt.fit(X_train, y_train)
R2_dt_tt= r2_score(y, dt_tt.predict(X))
print("R2 Decision Tree con train-test :",R2_dt_tt) 
```

### Repetir con Polynomial degree=2

```python
# Repetir con Polynomial degree=2
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Dividimos el dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_poly,y, 
    test_size=0.20,
    random_state=42,
)

# StandardScaler()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   
X_test_scaled = scaler.transform(X_test)      

lr_poly = LinearRegression()
lr_poly.fit(X_train_scaled, y_train)
R2_lr_poly_tt = r2_score(y_test, lr_poly.predict(X_test_scaled))
print("R2 Linear Regression poly con train-test:", R2_lr_poly_tt)
```

```python
# Decision Tree 
dt_tt = DecisionTreeRegressor()
dt_tt.fit(X_train, y_train)
R2_dt_poly_tt = r2_score(y_test, dt_tt.predict(X_test))
print("R2 Decision Tree poly con train-test:", R2_dt_poly_tt, 4)
```

```python
# Hacer un dataframe con los resultados obtenidos y compararlos
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Linear Regression con train test', 'Decision Tree con train test','Linear Regression Poly con train test', 'Decision Tree Poly con train test'],
    'R2 Score': [R2_lr, R2_dt, R2_lr_tt, R2_dt_tt,R2_lr_poly_tt,R2_dt_poly_tt]
})
results
```
